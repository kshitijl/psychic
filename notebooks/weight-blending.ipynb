```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K, L, M = 15, 2, 2
x = np.linspace(0, 200, 1000)
y = (1 + np.tanh(x/K - L)) / M

plt.plot(x, y, label="ml weight")

# vertical lines
for xv in [10, 30, 50, 100]:
    plt.axvline(xv, color='red', linestyle='--')
    yv = (1 + np.tanh(xv/K - L)) / M
    plt.scatter(xv, yv, color='black')
    plt.text(xv + 2, yv, f"({xv:.0f}, {yv:.3f})", va='bottom')

plt.xlim(0, 200)
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
